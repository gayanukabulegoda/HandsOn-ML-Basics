```python
import pandas as pd
df = pd.read_csv('advertising.csv')

# As per above context, we will use 'TV' as the independent variable and 'Sales' as the dependent variable.
# Set the independent variable (X) and dependent variable (y)
# Here, 'TV' is the feature and 'Sales' is the target variable.

# Set values for x and y
x = df[['TV']]
y = df['Sales']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42) 
# Split the data into training and testing sets
# Random state is set for reproducibility / training and testing sets

print("Training set (X_train):", X_train.shape)
print("Testing set (X_test):", X_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

# Train the model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Print the coefficients and intercept
print("Model coefficients and intercept:")
print("Coefficients:", model.coef_[0]) # Coefficients for the feature 'TV'
print("Intercept:", model.intercept_)
```

```python
import matplotlib.pyplot as plt

# Plot the regression line on the test data 
plt.figure(figsize=(8, 6)) # means 8 inches wide and 6 inches tall
plt.scatter(X_test, y_test, color='blue', label='Actual Sales') # scatter means actual data points with blue color
plt.plot(X_test, model.predict(X_test), color='red', linewidth=2, label='Predicted Sales (Regression Line)') # regression line in red color
plt.title('TV Ad Spending vs. Sales (Test Set)')
plt.xlabel('TV Ad Spending ($ thousands)')
plt.ylabel('Sales (thousands of units)')
plt.legend()
plt.show()
```

```python
from sklearn import metrics
import numpy as np

# Make predictions on the test set
y_pred_simple = model.predict(X_test)

# Calculate and print the evaluation metrics
mse_value = metrics.mean_squared_error(y_test, y_pred_simple)
print(f"Mean Squared Error (MSE): {mse_value:.4f}") # Here "mse_value: .4f" means it will print the value with 4 decimal places
print(type(mse_value))

# Calculate and print the Root Mean Squared Error (RMSE)
rmse_value = np.sqrt(mse_value)
print(f"Root Mean Squared Error (RMSE): {rmse_value:.4f}")
```
